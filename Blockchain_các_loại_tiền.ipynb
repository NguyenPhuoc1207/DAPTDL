```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import scipy
import statistics
import seaborn as sns 
from scipy import stats
   
```

### Thu thập dữ liệu và thực hiện thống kê mô tả

### Thu thập dữ liệu và thực hiện thống kê mô tả


```python
data = pd.read_csv("coin_Bitcoin.csv")
data.head(1000)
```

#### xem tổng quát dữ liệu


```python
sns.pairplot(data)
```

```python
# Tính trung bình các cột 

print("Giá trị trung bình của dataset:")
pd.DataFrame(data.mean()).T
```

```python
# Tính trung vị 

print("Trung vị của dataset:")
pd.DataFrame(data.median()).T
```

```python
# Tìm mode

print("Yếu vị của dataset:")
data.mode(dropna = True)
```

```python
min_value = data.min(skipna= True, numeric_only= True)
print(min_value)
print('-'*50)
```

```python
max_value = data.max(skipna= True, numeric_only= True)
print(max_value)
print('-'*50)
```

```python
print('Phạm vi của dữ liệu:')
print(max_value - min_value)
print('-'*50)
```

```python
# Tính độ lệch chuẩn - standard deviation
print("Độ lệch chuẩn của dataset:")
data.std(skipna = True)
```

#### Tương quan giữa giá cao và giá thấp trong sàn giao dịch

```python
plt.figure(figsize=(7, 10))
plt.title('Compare High with Low', fontsize=15)
data[['High', 'Low']].boxplot()
plt.ylabel('Count')
```

```python
#Biểu đổ thể hiện giá thấp nhất và giá cao nhất theo sàn giao dịch


data[["High", "Low"]].plot()
plt.ylabel('Count')
```
